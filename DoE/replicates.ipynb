```python
design_df = design_df.replace({'experience':mapping, 'other':mapping})
```

```python
design_df['price'].replace(15, 1,inplace=True)
design_df['price'].replace(20, 2,inplace=True)
design_df['price'].replace(25, 3,inplace=True)
design_df['price'].replace(30, 4,inplace=True)
```

```python
rep_design_df = pd.concat([design_df]*358).reset_index()
rep_design_df.shape
```

```python
samp_pop_df = pop_df.sample(n=6802, random_state=264).reset_index()
samp_pop_df.shape
```

```python
merged_df = pd.concat([samp_pop_df, rep_design_df], axis=1)
merged_df = merged_df.drop(['index','VAR1','LONG','LAT'], axis=1)
```
